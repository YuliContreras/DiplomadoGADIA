```python
#Operaciones basicas con cadenas
#La concatenacion
saludo="Hola"
nombre="Mundo"
mensaje=saludo+", "+nombre
print(mensaje)
```

```python
#La repeticion
repetir="hola"*3
print(repetir)
```

```python
#La indexacion: Acceder a caracteres individuales usando indices.
primer_caracter=mensaje[0]
print(primer_caracter)
```

```python
#Slicing: Obtiene subcadenas usando una notacion de rango
#Obtener la palabra "Hola" de la variable mensaje
sub_cadena=mensaje[0:4]
print(sub_cadena)
```

```python
cadena="python es genial"
#metodos de cadenas
#MAYUSCULAS
print(cadena.upper())
#minusculas
print(cadena.lower())
#tipo titulo
print(cadena.strip())
#remplazo de caracteres
print(cadena.replace("genial","increible"))
#separacion de cadenas
print(cadena.split(" "))
#union de cadenas
print(",".join(["python es genial"]))
```

```python
#Listas: """
"""
Las listas en Python son una de las estructuras de datos m谩s utilizadas debido a su flexibilidad y capacidad para almacenar diferentes tipos de datos. Aqu铆 te explico en detalle c贸mo funcionan las listas y c贸mo puedes utilizarlas"""
```

```python
#Creando una lista basica
mi_lista=[]

#Crear lista con elementos
mi_lista=[1,2,3,"cuatro",5.0,True]

#Accediendo a elementos de la lista
mi_lista = [10, 20, 30, 40, 50]

#Acceder al tercer elemento
print(mi_lista[2])

#Acceder al ultimo elemento
print(mi_lista[-1])
```

```python
#Modificacion de listas
#Cambian  en mi lista el 30 por 1
mi_lista[2]=1
print(mi_lista)
```

```python
#Agregar nuevos elementos a una lista
mi_lista=[1,2,3]

#Agregar un elemento al final de la lista
mi_lista.append(4)

print(mi_lista)
```

```python
#Insertar un elemento en la lista en una posicion especifica
mi_lista.insert(1,1.5)
print(mi_lista)
```

```python
#Agregar varios elementos a una lista
mi_lista.extend([5,6])
print(mi_lista)
```

```python
#eliminar elementos de las listas
mi_lista.remove(1)
print(mi_lista)
```

```python
#Eliminacion de elementos por una posicion especifica
del mi_lista[0]
print(mi_lista)
```
